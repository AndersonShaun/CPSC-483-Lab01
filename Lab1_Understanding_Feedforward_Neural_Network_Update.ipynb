```python
# model 1

from tensorflow import keras
from keras import layers
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout, Activation, Flatten
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(xc, yc, test_size=0.4, random_state=None)

model = Sequential()
model.add(Input(shape=(1,)))
model.add(Dense(150,activation = "relu"))
model.add(Dense(150,activation = "LeakyReLU"))
model.add(Dense(150,activation = "ThresholdedReLU"))
model.add(Dense(150,activation = "PReLU"))
model.add(Dense(150,activation = "ELU"))
model.add(Dense(150,activation = "relu"))
model.add(Dense(1))

model.compile(
   loss = 'MAE', optimizer = 'Adam', metrics=['mae']
)
model.fit(X_train, Y_train, batch_size = 128, epochs = 10)
```

```python
# model 3

from tensorflow import keras
from keras import layers
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout, Activation, Flatten
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(xc, yc, test_size=0.4, random_state=None)

model = Sequential()
model.add(Input(shape=(1,)))
model.add(Dense(100,activation = "relu"))
model.add(Dense(20,activation = "LeakyReLU"))
model.add(Dense(20,activation = "ThresholdedReLU"))
model.add(Dense(20,activation = "PReLU"))
model.add(Dense(20,activation = "ELU"))
model.add(Dense(20,activation = "relu"))
model.add(Dense(1))

model.compile(
   loss = 'MAE', optimizer = 'Adam', metrics=['mae']
)
model.fit(X_train, Y_train, batch_size = 128, epochs = 20)
```
